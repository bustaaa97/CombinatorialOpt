```python
def graphto0 (G):

    for i in G:
        G.nodes[i]["quemado"] = 0
        G.nodes[i]["defendido"] = 0

    return G
```

```python
def algorithm1 (G):
    for i in range (G.order()):
        if (list(nx.degree(G))[i][1]>3):
            return 0
        
    for i in G.nodes:
        G.nodes[i]["camino"] = 0
    
    r=random.choice(list(nx.degree(G)))
    while(r[1]==3):
        r=random.choice(list(nx.degree(G))) 
    #r = ("Segovia",2) #imponemos r  para las pruebas
    for i in G:
        if i == r[0]: continue
        if nx.degree(G,i)<3:
            G.nodes[i]["camino"] = nx.shortest_path_length(G,r[0],i)+1
        else:
            c=999
            for p in G:
                l=nx.cycle_basis(G,p)
                for n in l:
                    if (i in n and len(n)<c):
                        c=len(n)        
            G.nodes[i]["camino"]=c+nx.shortest_path_length(G,r[0],i)-1        
            
    print("Fire starts in "+r[0])        
    return r[0]
```

```python
def search_min(G):
    listcam=[]
    for n in G:
        listcam.append((G.nodes[n]["camino"],n))
    sorted(listcam)
    return listcam[1][1]
```

```python
#NO USE

def seleccionar_p (G, inicio):
    
    p = -1
    
    for i in G.nodes:
        
        if i != inicio:     
            if p == -1: p = i
            elif G.nodes[i]["camino"] < G.nodes[p]["camino"]: p = i
    
    return p
```

```python
def defense (G,inicio,path):
    
    d = -1
    c = -1
    
    for nbr in E.neighbors(inicio):
        if G.nodes[nbr]["quemado"] == 0 and G.nodes[nbr]["defendido"] == 0:
            c = nbr
            if nbr not in path:
                d = nbr
                
    if d == -1 and c != 1: d = c
        
    return d
```

```python
#path = list(nx.all_shortest_paths(G,inicio,p))[0]

def firefighter (G, inicio, path, t):
    
    G.nodes[inicio]["quemado"] = 1
    
    d = defense(G,inicio,path)
    
    if d != -1:
        
        G.nodes[d]["defendido"] = 1
        
        for nbr in E.neighbors(inicio): 
            
            if G.nodes[nbr]["quemado"] == 0 and G.nodes[nbr]["defendido"] == 0:
                t =firefighter (G, nbr, path, t+1)
                
    
    return t
```

```python
def alternatives(G,inicio,t):
    alts=[]
    for n in G:
        if n==inicio: continue
        if nx.shortest_path_length(G,inicio,n) <= (t+2):
            alts.append(n)
    return alts        
```

```python
def alts_chosen(E,alts):
    alts_correct=[]
    for i in alts:
        E=graphto0(E)
        z=firefighter(E,inicio_r,list(nx.all_shortest_paths(E,inicio_r,i))[0],0) 
        if z == nx.shortest_path_length(E,inicio_r,i):
            alts_correct.append(i)
    return(alts_correct)        
```

```python
def count(E,inicio,cities):
    total=[]
    for i in cities:
        t1=0
        t2=0
        for j in list(nx.all_shortest_paths(E,inicio,i))[0]:
            t1+=E.nodes[j]["population"]
            t2+=E.nodes[j]["surface"]
        total.append((i,round(t1,4),round(t2,4)))
        
    return total        
```
